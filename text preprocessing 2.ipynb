```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import re
nltk.download('punkt')
nltk.download('stopwords')
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
corpus = [ "I am loving the NLP class, but sometimes it feels confusing!!!",
          "NLP is a fascinating field — it deals with text, speech, and language understanding."]
print("My corpus : \n",corpus)
```

```python
vectorizer = CountVectorizer()
```

```python
bow_matrix = vectorizer.fit_transform(corpus)
print("bag of words Matriz:\n", bow_matrix.toarray())
```

```python
vocab = vectorizer.get_feature_names_out()
print("vocabulary:\n",vocab)
```
